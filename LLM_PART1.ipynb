```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    # Use separators instead of separator
    separators = ["\n"],
    chunk_size=200,
    chunk_overlap=0
)
```

```python
chunks = splitter.split_text(text)
len(chunks)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " "],  # List of separators based on requirement (defaults to ["\n\n", "\n", " "])
    chunk_size = 200,  # size of each chunk created
    chunk_overlap  = 0,  # size of  overlap between chunks in order to maintain the context
    length_function = len  # Function to calculate size, currently we are using "len" which denotes length of string however you can pass any token counter)
)
chunks=r_splitter.split_text(text)
len(chunks)
```

```python
for chunk in chunks:
  print (len(chunk))
```

```python
from qdrant_client.models import VectorParams, Distance

qdrant_client.recreate_collection(
    collection_name="my_text_collection",
    vectors_config=VectorParams(size=dim, distance=Distance.COSINE)
)
```

```python
from qdrant_client.models import PointStruct

# Create PointStructs
points = [
    PointStruct(id=i, vector=vectors[i], payload={"text": df.text.iloc[i]})
    for i in range(len(df))
]

# Upload to Qdrant
qdrant_client.upsert(
    collection_name="my_text_collection",
    points=points
)
```

```python
#query = "when does the concert starts?"
query = input("Enter your search query: ")

query_vector = encoder.encode(query).tolist()

# Perform search
results = qdrant_client.search(
    collection_name="my_text_collection",
    query_vector=query_vector,
    limit=3
)

# Display results
for result in results:
    print(f"Score: {result.score:.3f} | Text: {result.payload['text']}")
```
